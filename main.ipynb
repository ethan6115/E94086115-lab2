```python
import pygame
import time
from pygame import font

WIN_WIDTH = 1024
WIN_HEIGHT = 600
BTN_WIDTH = 80
BTN_HEIGHT = 80
HP_WIDTH = 40
HP_HEIGHT = 40
ENEMY_WIDTH=60        #設定敵人大小
ENEMY_HEIGHT=60
FPS = 30

# color (RGB)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# initialization
pygame.init()
# load image (background, enemy, buttons)
background_image = pygame.transform.scale(pygame.image.load("images/Map.png"), (WIN_WIDTH, WIN_HEIGHT))        
enemy_image = pygame.transform.scale(pygame.image.load("images/enemy.png"), (ENEMY_WIDTH, ENEMY_HEIGHT))      #輸入敵人圖片
hp_image = pygame.transform.scale(pygame.image.load("images/hp.png"), (HP_WIDTH, HP_HEIGHT))                  #輸入血量圖片
hpgray_image = pygame.transform.scale(pygame.image.load("images/hp_gray.png"), (HP_WIDTH, HP_HEIGHT))         #輸入灰色血量圖片
muse_image = pygame.transform.scale(pygame.image.load("images/muse.png"), (BTN_WIDTH, BTN_HEIGHT))            #輸入靜音圖片
sound_image = pygame.transform.scale(pygame.image.load("images/sound.png"), (BTN_WIDTH, BTN_HEIGHT))          #輸入音量圖片
continue_image = pygame.transform.scale(pygame.image.load("images/continue.png"), (BTN_WIDTH, BTN_HEIGHT))    #輸入繼續圖片
pause_image = pygame.transform.scale(pygame.image.load("images/pause.png"), (BTN_WIDTH, BTN_HEIGHT))          #輸入暫停圖片

# set the title
win = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))       
pygame.display.set_caption("My first game")               #設定標題



class Game:
    def __init__(self):
        
        # hp
        self.hp = 7                                       
        self.max_hp = 10
        pass

    def game_run(self):
        # game loop
        start_time=time.time()                              #記錄程式開始時間
        run = True
        while run:
            # event loop
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    run = False

            # 印出背景
            win.blit(background_image, (0, 0))
            
            # 印出敵人和血條
            win.blit(enemy_image, (20, 250))
            pygame.draw.rect(win, (255, 0, 0), [20, 240, 60, 5])
            
            # draw menu (and buttons)
            pygame.draw.rect(win, (0, 0, 0), [0, 0, WIN_WIDTH, 80])         #印出上面的黑色部分
            for i in range(0,200,40):                       #用for迴圈印出第一排的hp
                win.blit(hp_image, (400+i, 0))
            for i in range(0,80,40):                        #用for迴圈印出第二排的綠色hp
                win.blit(hp_image, (400+i, 40))
            for i in range(0,120,40):                       #用for迴圈印出第二排的灰色hp
                win.blit(hpgray_image, (480+i, 40))
            win.blit(muse_image, (1024-BTN_WIDTH-BTN_WIDTH-BTN_WIDTH-BTN_WIDTH, 0))        #利用螢幕邊緣=1024減掉按鈕寬度的方式來指定按鈕的位置
            win.blit(sound_image, (1024-BTN_WIDTH-BTN_WIDTH-BTN_WIDTH, 0))
            win.blit(continue_image, (1024-BTN_WIDTH-BTN_WIDTH, 0))
            win.blit(pause_image, (1024-BTN_WIDTH, 0))
            
            # draw time
            
            minute=int((time.time()-start_time)/60)                             #紀錄經過分鐘數
            second=int((time.time()-start_time)%60)                             #紀錄經過秒數
            strsecond=str(second)
            font = pygame.font.SysFont("simhei", 50)                             #設定字形和文字內容並輸出文字
            text_surface = font.render(f'{minute}:{strsecond.zfill(2)}', True, (255,255,255), (0,0,0))     
            win.blit(text_surface, (0,570))

            pygame.display.update()

if __name__ == "__main__":
    covid_game = Game()
    covid_game.game_run()

pygame.quit()

```
